```python
import numpy as np
from qiskit import *
from qiskit import Aer
from qiskit.visualization import plot_histogram, plot_state_city
from qiskit_ibm_runtime import QiskitRuntimeService, Estimator, Options
```

```python
#TP grimoire Aika
qc = QuantumCircuit(2)
#on creer notre circuit quantique de 2 qbits
```

```python
#On créer le diffuseur
qc.h(0)
qc.h(1)
qc.z(0)
qc.z(1)
qc.cz(0,1)
qc.h(0)
qc.h(1)
```

```python
#On mesure la valeur des qbits afin d'avoir le résultat de l'enigme soit quel grimoire est le bon
qcMeasure = QuantumCircuit(2)
qcMeasure.measure_all()
```

```python
#On affiche le résultat sous forme d'histogram
backend = BasicAer.get_backend('qasm_simulator') 
qc = qc.compose(qcMeasure)
qcResult = backend.run(transpile(qc, backend), shots=1000).result()
qcCounts  = qcResult.get_counts(qc)
plot_histogram(qcCounts)
```

```python
#La reponse est donc que c'est le grimoire 10 qui est le bon
```
